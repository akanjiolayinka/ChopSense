# ChopSense — Recommendation Exploration

Quick look at the processed sample dataset produced by `data/pipeline.py`.
Run `python -m data.pipeline --all --source huggingface` (or `--source local --path ./data/raw`) first.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

SAMPLE_DIR = Path('../data/processed/sample')
reviews = pd.read_parquet(SAMPLE_DIR / 'sample_reviews.parquet')
businesses = pd.read_parquet(SAMPLE_DIR / 'sample_businesses.parquet')
users = pd.read_parquet(SAMPLE_DIR / 'sample_users.parquet')
len(reviews), len(businesses), len(users)
```

```python
print('Reviews:    ', len(reviews))
print('Businesses: ', len(businesses))
print('Users:      ', len(users))
print('Avg words / review:', reviews['word_count'].mean().round(1))
print('Avg reviews / user:', reviews.groupby('user_id').size().mean().round(1))
reviews['stars'].value_counts().sort_index()
```

```python
counts = reviews['stars'].value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.bar(counts.index.astype(int), counts.values)
plt.xlabel('Stars')
plt.ylabel('Review count')
plt.title('Rating distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(reviews['word_count'], bins=50)
plt.xlabel('Words per review')
plt.ylabel('Frequency')
plt.title('Review length distribution')
plt.tight_layout()
plt.show()
```
